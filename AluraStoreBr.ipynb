### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
faturamento_total = {
    'Loja 1': loja['Preço'].sum(),
    'Loja 2': loja2['Preço'].sum(),
    'Loja 3': loja3['Preço'].sum(),
    'Loja 4': loja4['Preço'].sum(),
}

print("Faturamento Total por Loja:\n")
for loja_nome, total in faturamento_total.items():
    print(f"{loja_nome}: R$ {total:,.2f}")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))
plt.bar(faturamento_total.keys(), faturamento_total.values(), color='lightgreen')
plt.title("Faturamento Total por Loja")
plt.xlabel("Lojas")
plt.ylabel("Faturamento (R$)")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

# 2. Vendas por Categoria

# 4. Produtos Mais e Menos Vendidos

```python
def exibir_vendas_extremas(df, nome_loja):
    contagem = df['Produto'].value_counts()
    mais_vendido = contagem.idxmax()
    menos_vendido = contagem.idxmin()
    print(f"{nome_loja}:")
    print(f" Produto mais vendido: {mais_vendido} ({contagem.max()} vendas)")
    print(f" Produto menos vendido: {menos_vendido} ({contagem.min()} vendas)\n")

exibir_vendas_extremas(loja, "Loja 1")
exibir_vendas_extremas(loja2, "Loja 2")
exibir_vendas_extremas(loja3, "Loja 3")
exibir_vendas_extremas(loja4, "Loja 4")
```

```python
import matplotlib.pyplot as plt
import numpy as np

def mais_menos_vendidos(df):
    contagem = df['Produto'].value_counts()
    return {
        'mais_produto': contagem.idxmax(),
        'mais_qtd': contagem.max(),
        'menos_produto': contagem.idxmin(),
        'menos_qtd': contagem.min()
    }

dados_loja1 = mais_menos_vendidos(loja)
dados_loja2 = mais_menos_vendidos(loja2)
dados_loja3 = mais_menos_vendidos(loja3)
dados_loja4 = mais_menos_vendidos(loja4)

lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

mais_produtos = [dados_loja1['mais_produto'], dados_loja2['mais_produto'], dados_loja3['mais_produto'], dados_loja4['mais_produto']]
mais_qtds = [dados_loja1['mais_qtd'], dados_loja2['mais_qtd'], dados_loja3['mais_qtd'], dados_loja4['mais_qtd']]

menos_produtos = [dados_loja1['menos_produto'], dados_loja2['menos_produto'], dados_loja3['menos_produto'], dados_loja4['menos_produto']]
menos_qtds = [dados_loja1['menos_qtd'], dados_loja2['menos_qtd'], dados_loja3['menos_qtd'], dados_loja4['menos_qtd']]

x = np.arange(len(lojas))

width = 0.35

fig, ax = plt.subplots(figsize=(12,6))

bars1 = ax.bar(x - width/2, mais_qtds, width, label='Mais Vendido', color='lightgreen')

bars2 = ax.bar(x + width/2, menos_qtds, width, label='Menos Vendido', color='pink')

ax.set_xticks(x)
ax.set_xticklabels(lojas)

ax.set_ylabel('Quantidade Vendida')
ax.set_title('Comparação dos Produtos Mais e Menos Vendidos por Loja')
ax.legend()

def rotular_barras(barras, nomes):
    for barra, nome in zip(barras, nomes):
        altura = barra.get_height()
        ax.annotate(
            nome,
            xy=(barra.get_x() + barra.get_width() / 2, altura),
            xytext=(0, 3),
            textcoords="offset points",
            ha='center', va='bottom', fontsize=9, rotation=45
        )

rotular_barras(bars1, mais_produtos)
rotular_barras(bars2, menos_produtos)

plt.show()
```

```python
import pandas as pd

urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"
]

lista_lojas = []
for i, url in enumerate(urls, 1):
    df = pd.read_csv(url)
    df['Loja'] = f'Loja {i}'
    lista_lojas.append(df)

df_lojas = pd.concat(lista_lojas, ignore_index=True)

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_lojas, x='lon', y='lat', hue='Loja', alpha=0.5)
plt.title('Localização das vendas por loja')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df_lojas, x='lon', y='lat', fill=True, cmap='Reds', thresh=0.1)
plt.title('Mapa de calor das vendas')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
